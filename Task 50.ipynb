```python
# Task 50: Final Challenge - Student Management System
# Problem: Create a system that manages student records with grades.

def add_student(students):
    name = input("Enter student name: ")
    if name in students:
        print("Student already exists.")
        return
    try:
        grade = float(input("Enter student grade: "))
        students[name] = grade
        print("Student added.")
    except ValueError:
        print("Invalid grade input.")

def view_students(students):
    if not students:
        print("No student records found.")
        return
    print("\nStudent Records:")
    for name, grade in students.items():
        print(f"{name}: {grade}")

def update_grade(students):
    name = input("Enter student name to update: ")
    if name not in students:
        print("Student not found.")
        return
    try:
        grade = float(input("Enter new grade: "))
        students[name] = grade
        print("Grade updated.")
    except ValueError:
        print("Invalid grade input.")

def delete_student(students):
    name = input("Enter student name to delete: ")
    if name in students:
        del students[name]
        print("Student deleted.")
    else:
        print("Student not found.")

def show_statistics(students):
    if not students:
        print("No student records to show statistics.")
        return
    grades = list(students.values())
    avg = sum(grades) / len(grades)
    highest = max(grades)
    lowest = min(grades)
    print(f"Average grade: {avg:.2f}")
    print(f"Highest grade: {highest}")
    print(f"Lowest grade: {lowest}")

def student_management_system():
    students = {}
    while True:
        print("\nStudent Management System Menu:")
        print("1. Add Student")
        print("2. View Students")
        print("3. Update Grade")
        print("4. Delete Student")
        print("5. Show Statistics")
        print("6. Exit")
        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            add_student(students)
        elif choice == '2':
            view_students(students)
        elif choice == '3':
            update_grade(students)
        elif choice == '4':
            delete_student(students)
        elif choice == '5':
            show_statistics(students)
        elif choice == '6':
            print("Exiting Student Management System.")
            break
        else:
            print("Invalid choice. Please try again.")

student_management_system()
```
